**Task 2:Build Advanced Preprocessing Function**


```python
import re

def preprocess_text(text):
    """
    Advanced NLP preprocessing function
    Returns:
        tokens (list), cleaned_sentence (string)
    """

    # Handle edge cases
    if text is None or str(text).strip() == "":
        return [], ""

    # 1. Convert to lowercase
    processed = text.lower()

    # 2. Remove URLs
    processed = re.sub(r"http\S+|www\S+", " ", processed)

    # 3. Remove email-like patterns
    processed = re.sub(r"\S+@\S+", " ", processed)

    # 4. Remove numbers
    processed = re.sub(r"\d+", " ", processed)

    # 5. Handle repeated characters (soooo → so)
    processed = re.sub(r"(.)\1{2,}", r"\1", processed)

    # 6. Remove special characters (keep only alphabets and space)
    processed = re.sub(r"[^a-z\s]", " ", processed)

    # 7. Remove extra spaces
    processed = re.sub(r"\s+", " ", processed).strip()

    # 8. Tokenization
    words = processed.split()

    # 9. Remove short tokens (≤2), keep "no", "not"
    final_tokens = []
    for word in words:
        if len(word) > 2 or word in ("no", "not"):
            final_tokens.append(word)

    # 10. Reconstruct cleaned sentence
    cleaned_sentence = " ".join(final_tokens)

    return final_tokens, cleaned_sentence
```

```python
text = "WOW!!! I have 2 dogs!!! Visit http://example.com now!!!"
tokens, clean = preprocess_text(text)

print(tokens)
print(clean)
```

**Task 3: Stress Testing**


```python
# Test dataset (covers emojis, numbers, slang, repeated chars, URLs, mixed case)
test_data = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

print("----- NLP Preprocessing: Stress Test Results -----\n")

for i, sentence in enumerate(test_data, start=1):

    tokens, cleaned = preprocess_text(sentence)

    print(f"Sentence {i}:")
    print("Original Text   :", sentence)
    print("Cleaned Tokens  :", tokens)
    print("Cleaned Sentence:", cleaned)
    print("-" * 60)
```

**Task 4: Token Analytics**

```python
def compute_token_stats(token_list):

    # Handle empty tokens
    if not token_list:
        return 0, 0, 0

    total_tokens = len(token_list)
    unique_tokens = len(set(token_list))
    avg_length = sum(len(word) for word in token_list) / total_tokens

    return total_tokens, unique_tokens, round(avg_length, 2)


print("----- Token Analytics -----\n")

for idx, sentence in enumerate(test_data, 1):

    tokens, _ = preprocess_text(sentence)
    total, unique, avg_len = compute_token_stats(tokens)

    print(f"Sentence {idx}: {sentence}")
    print(f"Total Tokens      : {total}")
    print(f"Unique Tokens     : {unique}")
    print(f"Avg Token Length  : {avg_len}")
    print("-" * 60)
```

**Task 5: Frequency Analysis**

```python
from collections import Counter

# Step 1: Collect all tokens from all sentences
all_words = []

for sentence in test_data:
    tokens, _ = preprocess_text(sentence)
    all_words.extend(tokens)

# Step 2: Create frequency counter
word_freq = Counter(all_words)

print("----- Frequency Analysis -----\n")

# Top 10 most frequent words
print("Top 10 Most Frequent Words:\n")
for word, count in word_freq.most_common(10):
    print(f"{word} : {count}")

# Top 5 least frequent words
print("\nTop 5 Least Frequent Words:\n")

# Convert to list and pick last 5 elements
least_common = list(word_freq.items())[-5:]

for word, count in least_common:
    print(f"{word} : {count}")
```

**Task 6: Build Full Pipeline**


```python
def full_pipeline(text_list):
    """
    Runs preprocessing on a list of sentences
    Returns:
        {
            "tokens": [...],
            "clean_sentences": [...]
        }
    """

    # Initialize storage
    aggregated_tokens = []
    cleaned_outputs = []

    # Process each sentence
    for item in text_list:
        tokens, cleaned_text = preprocess_text(item)

        # Collect results
        aggregated_tokens.extend(tokens)
        cleaned_outputs.append(cleaned_text)

    # Final structured output
    result = {
        "tokens": aggregated_tokens,
        "clean_sentences": cleaned_outputs
    }

    return result
```

```python
pipeline_output = full_pipeline(test_data)

print("----- Full Pipeline Output -----\n")

print("All Tokens:\n")
print(pipeline_output["tokens"])

print("\nCleaned Sentences:\n")
for i, sent in enumerate(pipeline_output["clean_sentences"], 1):
    print(f"{i}. {sent}")
```
